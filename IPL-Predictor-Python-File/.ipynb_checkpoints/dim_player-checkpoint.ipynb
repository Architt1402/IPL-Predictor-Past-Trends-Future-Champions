```python
import pandas as pd

match_dates = dim_df['matchDate'].astype(str)

# Masks
slashes = match_dates.str.contains("/", na=False)
dashes_4digit = match_dates.str.match(r'\d{2}-\d{2}-\d{4}', na=False)
month_short = match_dates.str.match(r'\d{2}-[A-Za-z]{3}-\d{2}', na=False)

# Parse each format
dates_slash = pd.to_datetime(match_dates[slashes], format='%m/%d/%Y', errors='coerce')
dates_dash4 = pd.to_datetime(match_dates[dashes_4digit], format='%m-%d-%Y', errors='coerce')
dates_month = pd.to_datetime(match_dates[month_short], format='%d-%b-%y', errors='coerce')

# Combine all parsed dates
parsed_dates = pd.concat([dates_slash, dates_dash4, dates_month]).sort_index()

# Assign cleaned and formatted dates back
dim_df.loc[parsed_dates.index, 'matchDate'] = parsed_dates.dt.strftime('%d/%m/%Y')
```

```python
# Mapping of short or alternate names to full team names
team_name_map = {
    'KKR': 'Kolkata Knight Riders',
    'RCB': 'Royal Challengers Bangalore',
    'mi': 'Mumbai Indians',
    'MI': 'Mumbai Indians',
    'Mumbai':'Mumbai Indians',
    'Capitals': 'Delhi Capitals',
    'DC': 'Delhi Capitals',
    'Punjab Kings': 'Punjab Kings',
    'Super Kings': 'Chennai Super Kings',
    'CSK': 'Chennai Super Kings',
    'Royals': 'Rajasthan Royals',
    'RR': 'Rajasthan Royals',
    'Sunrisers': 'Sunrisers Hyderabad',
    'SRH': 'Sunrisers Hyderabad',
    'Titans': 'Gujarat Titans',
    'GT': 'Gujarat Titans',
    'Super Giants': 'Lucknow Super Giants',
    'LSG': 'Lucknow Super Giants'
}

# Apply mapping to all team columns
dim_df['team1'] = dim_df['team1'].replace(team_name_map)
dim_df['team2'] = dim_df['team2'].replace(team_name_map)
dim_df['winner'] = dim_df['winner'].replace(team_name_map)
```

```python
dim_df['year'] = dim_df['matchDate'].dt.year
filtered_df = dim_df[dim_df['year'].isin([2022, 2023, 2024])]
print(dim_df.dtypes)

```

```python
matches_per_year_played = filtered_df.groupby('year').size()
print(matches_per_year_played)
```

```python
wins_per_team_year = filtered_df.groupby(['year', 'winner']).size().unstack(fill_value=0)
print(wins_per_team_year)
```

```python
total_wins_per_team = filtered_df['winner'].value_counts()
print(total_wins_per_team)
```

```python
# Count appearances in 'team1' and 'team2' for each year
team1_counts = filtered_df.groupby(['year', 'team1']).size()
team2_counts = filtered_df.groupby(['year', 'team2']).size()

# Combine counts
matches_played = team1_counts.add(team2_counts, fill_value=0).unstack(fill_value=0)
print(matches_played)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

matches_long = matches_played.reset_index().melt(id_vars='year', var_name='Team', value_name='Matches')

plt.figure(figsize=(14, 7))
sns.barplot(data=matches_long, x='year', y='Matches', hue='Team')
plt.title('Matches Played Per Team Per Year')
plt.xlabel('Year')
plt.ylabel('Matches Played')
plt.legend(title='Team', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
wins_per_team_year = filtered_df.groupby(['year', 'winner']).size().unstack(fill_value=0)

wins_per_team_year.plot(kind='bar', figsize=(12, 6))
plt.title('Number of Wins per Team per Year')
plt.xlabel('Year')
plt.ylabel('Number of Wins')
plt.xticks(rotation=0)
plt.legend(title='Team', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
total_wins_per_team.sort_values().plot(kind='barh', color='green')
plt.title('Total Wins per Team (2022–2024)')
plt.xlabel('Number of Wins')
plt.ylabel('Team')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a pivot table for the heatmap
heatmap_data = matches_played.T

# Plot the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', cbar=True, fmt="d")
plt.title('Matches Played per Team per Year )')
plt.xlabel('Year')
plt.ylabel('Team')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
